```python
import requests
from requests_oauthlib import OAuth1
```

```python
yelp_consumer_key    = 'XXXXXXXXXX';
yelp_consumer_secret = 'XXXXXXXXXX';
yelp_token           = 'XXXXXXXXXX';
yelp_token_secret    = 'XXXXXXXXXX';
```

```python
# Let's combine this together into a function

def yelp_search(term="Food", location="Newport Beach"):
    base_url = "https://api.yelp.com/v2/search"
    params = {"term": "Food", "location": "New Port Beach"}
    auth = OAuth1(yelp_consumer_key, yelp_consumer_secret, yelp_token, yelp_token_secret)
    r = requests.get(base_url, auth=auth, params=params)
    return r.json()
```

```python
search_1 = yelp_search()

for value in search_1["businesses"]:
    print(value["name"])
    print(value["phone"])
    print(value.get("website")) # .get returns none if key is not found
    print(value.get("image_url"))
    print(value["location"]["display_address"])
    print(value["location"]["city"])
    print(value.get("location").get("area"))
    print("\n")
```
